```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=filtered_frequency_table.index, y=filtered_frequency_table.values, palette="Set3")
plt.xlabel('Categories')  # Set x-axis label
plt.ylabel('Frequency')   # Set y-axis label
plt.title('Barplot of Frequency Table of Top 20 Product Categories (excl. Sport & Outdoors)')  # Add a title
plt.xticks(rotation=90)  # Rotate x-axis labels if necessary
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Get value counts of the 'category' column
category_counts = proportion_table

# Step 2: Create the pie chart
plt.pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%', startangle=50)

# Step 3: Customize and show the chart
plt.title('Category Distribution')
plt.axis('equal')  # Ensures that the pie is drawn as a circle
plt.show()
```

```python
category_counts = filtered_proportion_table

# Step 2: Create the pie chart
plt.pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%', startangle=50)

# Step 3: Customize and show the chart
plt.title('Category Distribution excluding Sports & Outdoors')
plt.axis('equal')  # Ensures that the pie is drawn as a circle
plt.show()
```

```python
#Determine the variance, standard deviation, range, and interquartile range for product price.
#How varied are the product prices? Are there any indicators of a significant spread in prices?
```

```python
var = round(df['price'].var(), 2)
var
```

```python
std_dev = round(df['price'].std(), 2)
std_dev
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
IQR
```
